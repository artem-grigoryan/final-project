```python
import requests
import pandas as pd
from datetime import datetime, timedelta
import folium
import networkx as nx
import matplotlib.pyplot as plt
import sqlite3
```

# 1. API

Извлечение информации с HeadHunter API. При указании параметров пагинации (page, per_page) работает ограничение: глубина возвращаемых результатов не может быть больше 2000, поэтому мне пришлось обходить данное ограничение.Также API HH.ru по умолчанию возвращает только 20 вакансий на страницу.  

```python
def get_professional_roles():  # Функция для получения справочника профессиональных ролей
    url = 'https://api.hh.ru/professional_roles'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        roles = {}
        for category in data['categories']:
            for role in category['roles']:
                roles[role['id']] = {
                    'name': role['name'],
                    'category': category['name']
                }
        return roles
    else:
        print(f"Ошибка получения справочника профессиональных ролей: {response.status_code}")
        return {}


def fetch_vacancies(start_date, end_date): # Функция для получения вакансий за указанный интервал дат
    params = {
        'area': 1,  # Код региона (1 - Москва)
        'per_page': 100,  # Количество вакансий на одной странице
        'page': 0,  # Номер страницы
        'date_from': start_date,
        'date_to': end_date
    }

    url = 'https://api.hh.ru/vacancies'
    vacancies = []

    while True:
        response = requests.get(url, params=params)

        if response.status_code == 200:
            data = response.json()
            vacancies.extend(data['items'])
            if params['page'] < data['pages'] - 1:
                params['page'] += 1
            else:
                break
        else:
            print(f"Ошибка: {response.status_code}")
            break

    return vacancies

roles_dict = get_professional_roles()


all_vacancies = []   # Получаем вакансии за разные интервалы дат
end_date = datetime.now()
start_date = end_date - timedelta(days=60)
interval = timedelta(days=10)  # Интервал в 10 дней

while start_date < end_date:
    interval_end = start_date + interval
    if interval_end > end_date:
        interval_end = end_date

    vacancies = fetch_vacancies(start_date.isoformat(), interval_end.isoformat())
    all_vacancies.extend(vacancies)

    start_date = interval_end


processed_vacancies = []    # Обрабатываем вакансии и извлекаем нужные данные
for vacancy in all_vacancies:
    role_names = []      # Получаем список профессиональных ролей и их категорий
    categories = []
    for role in vacancy.get('professional_roles', []):
        role_id = role['id']
        role_info = roles_dict.get(role_id, {'name': 'Неизвестная роль', 'category': 'Неизвестная категория'})
        role_names.append(role_info['name'])
        categories.append(role_info['category'])

    processed_vacancy = {
        'Название': vacancy['name'],
        'Компания': vacancy['employer']['name'],
        'Ссылка': vacancy['alternate_url'],
        'Профессиональные роли': ', '.join(role_names),
        'Категории': ', '.join(categories),
    }

    address = vacancy.get('address')    # Проверка наличия адреса
    if address:
        processed_vacancy['Адрес'] = ', '.join(filter(None, [address.get('city'), address.get('street'), address.get('building')]))
        processed_vacancy['Станции метро'] = ', '.join([metro['station_name'] for metro in address.get('metro_stations', [])])
        processed_vacancy['lat'] = address.get('lat')
        processed_vacancy['lng'] = address.get('lng')
    else:
        processed_vacancy['Адрес'] = None
        processed_vacancy['Станции метро'] = None
        processed_vacancy['lat'] = None
        processed_vacancy['lng'] = None

    if vacancy['salary']:    # Проверяем наличие информации о зарплате
        processed_vacancy['Зарплата от'] = vacancy['salary']['from']
        processed_vacancy['Зарплата до'] = vacancy['salary']['to']
        processed_vacancy['Валюта'] = vacancy['salary']['currency']
    else:
        processed_vacancy['Зарплата от'] = None
        processed_vacancy['Зарплата до'] = None
        processed_vacancy['Валюта'] = None

    processed_vacancies.append(processed_vacancy)


df = pd.DataFrame(processed_vacancies)   # Создаем DataFrame и сохраняем в CSV файл
csv_file_path = 'vacancies.csv'
df.to_csv(csv_file_path, index=False)

print(f"Найдено {len(all_vacancies)} вакансий")
print(f"Данные сохранены в файл {csv_file_path}")
```

# 2. Гистограмма распределения средней зарплаты по категориям

Далее рассмотрим среднюю зарплату в каждой из категориях для дальнейшего анализа (и визуализации!).

```python
df = pd.read_csv('vacancies_with_average_salary.csv')

df = df.dropna(subset=['Средняя зарплата'])  # Удаление вакансий со значением средней зарплаты NaN

average_salaries_by_category = df.groupby('Категории')['Средняя зарплата'].mean().reset_index()  # Группировка по категориям и вычисление среднего значения зарплаты

average_salaries_by_category.columns = ['Категория', 'Средняя зарплата']


average_salaries_by_category.to_csv('average_salaries_by_category.csv', index=False)   # Сохранение результатов в новый CSV файл

print(f"Средние зарплаты по категориям сохранены в файл 'average_salaries_by_category.csv'")

print(average_salaries_by_category)
```

```python
category_salaries = df.groupby('Категории')['Средняя зарплата'].mean().reset_index()  # Группировка по категориям и вычисление средней зарплаты


category_salaries = category_salaries.sort_values(by='Средняя зарплата', ascending=False) # Сортировка по средней зарплате в порядке убывания


category_salaries['Средняя зарплата'] = category_salaries['Средняя зарплата'].round()  # Округление средней зарплаты до целых значений

plt.figure(figsize=(21, 11))  # Построение графика

bars = plt.bar(category_salaries['Категории'], category_salaries['Средняя зарплата'])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f"{int(height)}",
             ha='center', va='bottom')

plt.xlabel('Категории', fontsize=14)
plt.ylabel('Средняя зарплата', fontsize=14)
plt.title('Распределение средней зарплаты по категориям', weight='bold', fontsize=15)

plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--')
plt.gca().set_facecolor('white')

plt.tight_layout()
plt.show()
```

Данное распределение показывает, что категория "Высший и средний менеджмент" имеет самую высокую средную заработную плату. И именно поэтому дальше мы будем делать акцент на этой категории.

# 3. Визуализация вакансий на интерактивной карте Москвы

Мы возьмем только категорию "Высший и средний менеджмент", так как 6000+ вакансий слишком много для качественной визуализации на карте.

```python

df = pd.read_csv('vacancies.csv')


filtered_df = df[df['Категории'].str.contains('Высший и средний менеджмент', na=False)]   # Фильтрация вакансий по категории "Высший и средний менеджмент"

moscow_map = folium.Map(location=[55.7558, 37.6173], zoom_start=10) # Создание базовой карты с центром в Москве

for index, row in filtered_df.iterrows():   # Добавление маркеров для каждой вакансии
    if not pd.isna(row['lat']) and not pd.isna(row['lng']):
        folium.Marker(
            location=[row['lat'], row['lng']],
            popup=f"<b>{row['Название']}</b><br>{row['Компания']}<br>Зарплата от: {row['Зарплата от']} до: {row['Зарплата до']} {row['Валюта']}",
            tooltip=row['Название']
        ).add_to(moscow_map)

moscow_map.save('moscow_vacancies_map.html')  # Сохранение карты в HTML файл
```

# 4. Граф компаний и профессиональных ролей

```python
selected_categories = ['Высший и средний менеджмент']  # Ограничиваем выборку вакансий
filtered_df = df[df['Категории'].isin(selected_categories)]


G = nx.Graph() # Создаем пустой граф


companies = filtered_df['Компания'].unique() # Добавляем узлы компаний
for company in companies:
    G.add_node(company, node_type='company')


roles = filtered_df['Профессиональные роли'].str.split(', ').explode().unique() # Добавляем узлы профессиональных ролей
for role in roles:
    G.add_node(role, node_type='role')

for _, row in filtered_df.iterrows():  # Добавляем ребра между компаниями и профессиональными ролями
    company = row['Компания']
    roles = row['Профессиональные роли'].split(', ')
    for role in roles:
        G.add_edge(company, role)


plt.figure(figsize=(20, 10)) # Рисуем граф
pos = nx.spring_layout(G, k=0.15, iterations=20)
node_colors = {'company': 'skyblue', 'role': 'lightgreen'}
node_types = nx.get_node_attributes(G, 'node_type')
nx.draw(G, pos, with_labels=True, node_size=1500, font_size=15, font_weight='bold',
        node_color=[node_colors[node_types[node]] for node in G.nodes()], edge_color='gray')
plt.title('Граф компаний и профессиональных ролей (Высший и средний менеджмент)')
plt.show()

```

В результате получается граф, который наглядно показывает связи между компаниями и профессиональными ролями в категории "Высший и средний менеджмент". Такая визуализация может быть полезна для анализа потребностей компаний в специалистах определенных ролей, а также для выявления трендов и паттернов на рынке труда в этой категории.

# 5. SQL

Поработаем с SQL, чтобы показать, что можем!

```python
conn = sqlite3.connect('vacancies.db')  # Создаем подключение к базе данных SQLite

df.to_sql('vacancies', conn, if_exists='replace', index=False)  # Сохраняем DataFrame в базе данных SQLite


salary_data = pd.read_sql_query("SELECT `Зарплата от`, `Зарплата до` FROM vacancies WHERE `Зарплата от` IS NOT NULL AND `Зарплата до` IS NOT NULL", conn)  # Загружаем данные о зарплате из базы данных


salary_data['Средняя зарплата'] = salary_data[['Зарплата от', 'Зарплата до']].mean(axis=1)


plt.hist(salary_data['Средняя зарплата'], bins=20, color='skyblue', edgecolor='black')  # Создаем гистограмму распределения средней зарплаты
plt.title('Распределение средней зарплаты')
plt.xlabel('Средняя зарплата')
plt.ylabel('Количество вакансий')
plt.show()
```

Гистограмма показывает распределение средних зарплат по всем вакансиям. Цель использования гистограммы в данном случае - получить визуальное представление о распределении зарплат в вакансиях. Это может помочь быстро увидеть, какие уровни зарплаты наиболее распространены, и определить, есть ли какие-то аномалии или особенности в данных.
